```r
hfun_logcondens <- function(X_samps){
    density_ratio <- evaluate_logcondens_estimator_vectorized(X_samps, my_estimator_logcondens)/heteroskedastic_gaussian_pdf_vectorized(X_samps, SimData)
    return(0.5*(sqrt(density_ratio)-1)^2)
}

hfun_logconcdead <- function(X_samps){
    density_ratio <- evaluate_logconcdead_estimator_vectorized(X_samps, my_estimator_logconcdead)/heteroskedastic_gaussian_pdf_vectorized(X_samps, SimData)
    return(0.5*(sqrt(density_ratio)-1)^2)
}

hfun_fulldim <- function(X_samps){
    density_ratio <- evaluate_fulldim_estimator_vectorized(X_samps, fulldim_output$my_fulldim_estimator)/heteroskedastic_gaussian_pdf_vectorized(X_samps, SimData)
    return(0.5*(sqrt(density_ratio)-1)^2)
}

generate_heteroskedastic_gaussian_samples_for_monte_carlo <- function(K_samps){
    return(mvrnorm(K_samps, mu=SimData$true_mean_vec, Sigma=SimData$covariance_X))
}
```

```r
hist(heteroskedastic_gaussian_pdf_vectorized(SimData$pre_data, SimData))
hist(evaluate_logcondens_estimator_vectorized(SimData$pre_data, my_estimator_logcondens))
```

```r

plot(x=heteroskedastic_gaussian_pdf_vectorized(SimData$pre_data, SimData), y=evaluate_logcondens_estimator_vectorized(SimData$pre_data, my_estimator_logcondens))
```

```r
K_samps <- 10000
num_repeats <- 5
hellinger_error_estimate_statistics_logcondens <- naive_monte_carlo_integrate_repeated(hfun_logcondens, 
                                                    generate_heteroskedastic_gaussian_samples_for_monte_carlo, K_samps, num_repeats)
# boxplot(hellinger_error_estimate_statistics_logcondens$all_vals)
```

```r
hellinger_error_estimate_statistics_logcondens$mean_val
```

```r
K_samps <- 10000
num_repeats <- 50
hellinger_error_estimate_statistics_fulldim <- naive_monte_carlo_integrate_repeated(hfun_fulldim, 
                                                    generate_heteroskedastic_gaussian_samples_for_monte_carlo, K_samps, num_repeats)
# boxplot(hellinger_error_estimate_statistics_logcondens$all_vals)
```

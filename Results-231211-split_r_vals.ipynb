```r


convert_to_matrix <- function(retrieved_data) {

    # Convert to matrix
    num_exp = length(retrieved_data)
    
    results_mat <- matrix(nrow=num_exp ,ncol=5)
    colnames(results_mat) <- c("n", "d", "split_r", "irep", "hell_err_mean_logcondens")
    
    for (eind in 1:num_exp){
        results_mat[eind, "n"] <- retrieved_data[[eind]]$n
        results_mat[eind, "d"] <- retrieved_data[[eind]]$d
        results_mat[eind, "split_r"] <- retrieved_data[[eind]]$split_r
        results_mat[eind, "irep"] <- retrieved_data[[eind]]$irep
        results_mat[eind, "hell_err_mean_logcondens"] <- retrieved_data[[eind]]$hell_err_mean_logcondens
    }
    return(results_mat)
}

get_ave_err <- function(results_mat, all_split_r_vals){
    num_r_vals <- length(all_split_r_vals)
    ave_err_vec <- rep(0,num_r_vals)
    for (rvind in 1:num_r_vals){
        where_r <- (results_mat[,"split_r"]==all_split_r_vals[rvind])
        ave_err_vec[rvind] <- mean(results_mat[where_r,"hell_err_mean_logcondens"])
    }
    return(ave_err_vec)
}
```

```r
# retrieved_data
retrieved_data_d5 <- readRDS('Results/split_r_exp_gaussian_d=5_.RData')
retrieved_data_d10 <- readRDS('Results/split_r_exp_gaussian_d=10_.RData')
retrieved_data_d15 <- readRDS('Results/split_r_exp_gaussian_d=15_.RData')



results_mat_d5 <- convert_to_matrix(retrieved_data_d5)
results_mat_d10 <- convert_to_matrix(retrieved_data_d10)
results_mat_d15 <- convert_to_matrix(retrieved_data_d15)

```

```r

pdf(file='Results/split-r-figures/all_three.pdf', width=15, height=5)
options(repr.plot.width = 15, repr.plot.height = 5)
par(mfrow = c(1, 3), mar=c(6,6,4,2))

point_alpha <- 0.3
line_alpha <- 1
pch <- 19
lty <- 1

plot(x=results_mat_d5[,"split_r"], results_mat_d5[,"hell_err_mean_logcondens"], pch=pch, col=rgb(0,0,1,point_alpha), main=expression(d == 5),
     cex.lab=2, cex.axis=1.5, cex.main=2.5, xlab=expression(r), ylab="Squared hellinger error", ylim=c(0,0.13))
lines(x=all_split_r_vals, y=get_ave_err(results_mat_d5, all_split_r_vals), lty=lty, col=rgb(0,0,1,line_alpha))
###
plot(x=results_mat_d10[,"split_r"], results_mat_d10[,"hell_err_mean_logcondens"], pch=pch, col=rgb(0,0,1,point_alpha), main=expression(d == 10),
     cex.lab=2, cex.axis=1.5, cex.main=2.5, xlab=expression(r), ylab="Squared hellinger error", ylim=c(0,0.13))
lines(x=all_split_r_vals, y=get_ave_err(results_mat_d10, all_split_r_vals), lty=lty, col=rgb(0,0,1,line_alpha))
###
plot(x=results_mat_d15[,"split_r"], results_mat_d15[,"hell_err_mean_logcondens"], pch=pch, col=rgb(0,0,1,point_alpha), main=expression(d == 15),
     cex.lab=2, cex.axis=1.5, cex.main=2.5, xlab=expression(r), ylab="Squared hellinger error", ylim=c(0,0.13))
lines(x=all_split_r_vals, y=get_ave_err(results_mat_d15, all_split_r_vals), lty=lty, col=rgb(0,0,1,line_alpha))
###
dev.off()
```
